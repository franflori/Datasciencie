```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Cargar el data frame


```python
df = pd.read_csv("..\..\datos\Iris.csv")
df
```

## Gráfico con  Pandas- petalo

```python
figura=df[df.Clase=="Iris-setosa"].plot(kind="scatter", x="PL", y ="PW",
color="blue", label='Setosa', figsize=(6.4, 6.4))
df[df.Clase=="Iris-versicolor"].plot(kind="scatter", x="PL", y ="PW",
color="red", label='Versicolor', ax=figura)
df[df.Clase=="Iris-virginica"].plot(kind="scatter", x="PL", y ="PW",
color="green", label='Virginica', ax=figura)

figura.set_xlabel("Longitud")
figura.set_ylabel("Anchura")
figura.set_title("Gráfica para el Pétalo - con Pandas")
plt.show()
```

## Gráfico con MATPLOTLIB petalo

```python
df_set=df[df.Clase=="Iris-setosa"]
df_ver=df[df.Clase=="Iris-versicolor"]
df_vir=df[df.Clase=="Iris-virginica"]

fig = plt.figure(figsize=(6.4, 6.4))
ax1= fig.add_subplot(111)
ax1.scatter(df_set["PL"], df_set["PW"], s=10, c="blue", label="Setosa")
ax1.scatter(df_ver["PL"], df_ver["PW"], s=10, c="red", label="Versicolor")
ax1.scatter(df_vir["PL"], df_vir["PW"], s=10, c="green", label="Virginica")

plt.legend(loc="upper left")
plt.title("Gráfica para el Pétalo - con Matplotlib")
plt.xlabel("Longitud")
plt.ylabel("Anchura")
plt.show()
```

## Gráfico con SEABORN petalo

```python
#pip install seaborn
```

```python
sns.FacetGrid(df, hue="Clase", height=6.4)\
.map(plt.scatter, "PL", "PW")\
.add_legend()
plt.title("Gráfica para el Pétalo - con Seaborn")
plt.show()
```
