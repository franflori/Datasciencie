```python
import plotnine as p9
```

```python
from plotnine import *
```

```python
import pandas as pd
```

```python
df = pd.read_csv('..\..\datos\iris.csv')
```

```python
# Crea un marco para el gráfico definiendo las variables
surveys_plot = p9.ggplot(data=df[df.Clase=="Iris-setosa"],
                         mapping=p9.aes(x='PL', y='PW')) + p9.geom_point(alpha=0.1, color='blue' ,figsize=(6.4, 6.4))
               




# Dibuja los puntos en el marco
surveys_plot 
```

```python
(p9.ggplot(data=df,
           mapping=p9.aes(x='PL', y='PW',
                          fill='Clase'))
    + p9.geom_bar()
    )
```

```python
ggplot(df) + aes(x="Clase", y="SL") + geom_col()
```

```python
(
    ggplot(df)
    + facet_grid(facets="~Clase")
    + aes(x="PL", y="PW")
    + labs(
        x="PL",
        y="PW",
    )
    + geom_col()
)
```
