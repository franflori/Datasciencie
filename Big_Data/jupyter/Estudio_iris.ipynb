```python
opciones = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm","PetalWidthCm"]
for opcion in opciones:
    pd.crosstab(df[opcion], df.Species).plot(kind="bar")
   
    plt.show()
```
