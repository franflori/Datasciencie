```python
########## VISUALIZAMOS LOS DATOS ##########

import matplotlib.pyplot as plt

#Grafico Sepal - Longitud vs Ancho
fig = iris[iris.Clase == 'Iris-setosa'].plot(kind='scatter',
          x='SL', y='SW', color='blue', label='Setosa')
iris[iris.Clase == 'Iris-versicolor'].plot(kind='scatter',
    x='SL', y='SW', color='green', label='Versicolor', ax=fig)
iris[iris.Clase == 'Iris-virginica'].plot(kind='scatter',
    x='SL', y='SW', color='red', label='Virginica', ax=fig)

fig.set_xlabel('Sépalo - Longitud')
fig.set_ylabel('Sépalo - Ancho')
fig.set_title('Sépalo - Longitud vs Ancho')
plt.show()

```

```python
#Grafico Pétalo - Longitud vs Ancho
fig = iris[iris.Clase == 'Iris-setosa'].plot(kind='scatter',
          x='PL', y='PW', color='blue', label='Setosa')
iris[iris.Clase == 'Iris-versicolor'].plot(kind='scatter',
    x='PL', y='PW', color='green',
    label='Versicolor', ax=fig)
iris[iris.Clase == 'Iris-virginica'].plot(kind='scatter',
    x='PL', y='PW', color='red', label='Virginica', ax=fig)

fig.set_xlabel('Pétalo - Longitud')
fig.set_ylabel('Pétalo - Ancho')
fig.set_title('Pétalo Longitud vs Ancho')
plt.show()
```

## Violin plot ##

```python
def VilonPlot(ejex,ejey,datos):
    sns.set(font_scale=1.8, rc={'figure.figsize':(12, 10)})
    sns.violinplot(x=ejex, y=ejey, data=datos, size=6)
    plt.show()
    
```

```python
sns.set(font_scale=1.8, rc={'figure.figsize':(12, 10)})
sns.violinplot(x="Clase", y="PL", data=iris, size=6)
plt.show()
```
