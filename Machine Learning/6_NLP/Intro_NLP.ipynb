 # 1.4 Palabras a eliminar


```python

palabras_eliminar = stopwords + signos_puntuacion
palabras_eliminar
```

```python
len(text1), len(palabras_eliminar)
```

Palabras importantes, en este caso solo sumo cuantas hay


```python
palabrasImportantes = 0
for palabra in text1:
    if palabra not in palabras_eliminar:
        palabrasImportantes += 1
palabrasImportantes
```

Palabras Importantes, aqui me creo la lista con ellas


```python
palabrasImportantesDB = []
for palabra in text1:
    if palabra not in palabras_eliminar:
        palabrasImportantesDB.append(palabra)
palabrasImportantesDB
```

```python
len(palabrasImportantesDB), len(text1)
```

Recordando la forma r√°pida de Python

```python
palabras_imp = []
[palabras_imp.append(palabra) for palabra in text1 if palabra not in palabras_eliminar]
palabras_imp
```

# 2.1 Counter

```python
dist2 = FreqDist(palabras_imp)
dist2
```

```python
dist2.plot(10)
plt.show()
```

Vamos a calcular con Counter

```python
from collections import Counter
```

```python
contadorPalabras = Counter(palabras_imp)
contadorPalabras
```

Ordeno de Mayor a Menor transformando el diccionario en dataframe

```python
listado_keys = []
for key in contadorPalabras.keys():
    listado_keys.append(key)
listado_keys
```

```python
listado_values = []
for value in contadorPalabras.values():
    listado_values.append(value)
listado_values
```

```python
import pandas as pd
df = pd.DataFrame({"Keys": listado_keys, "values": listado_values})
df.head()
```

```python
df_mayor_menor = df.sort_values(by="values", ascending=False)
df_mayor_menor.head()
```
