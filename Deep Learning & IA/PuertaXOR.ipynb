## Puerta XOR, Or Exclusiva

```python
import pandas as pd
df = pd.DataFrame({'Entrada_1': [0, 0, 1, 1],
'Entrada_2': [0, 1, 0, 1],
'Salida_12': [0, 1, 1, 0]})
```

```python
import numpy as np
```

```python
training_data = np.array([[0,0], [0,1], [1,0], [1,1]], "float32")
training_data
```

```python
target_data = np.array([[0], [1], [1], [0]], "float32")
target_data
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
training_data.shape # 4 filas 2 columnas (2 variables)
```

```python
len(training_data)
```

```python
model = Sequential()
```

```python
model.add(Dense(4, input_dim = 2, activation="relu"))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile(optimizer='adam',
loss = 'mean_squared_error',
metrics=['binary_accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(training_data, target_data, epochs=5000, verbose=2)
```

```python
model.predict(training_data)
```

```python
model.predict(training_data).round()
```

```python
# model.predict_classes(training_data)
# AttributeError: 'Sequential' object has no attribute 'predict_classes'
```

```python
predictions = (model.predict(training_data) > 0.5).astype("int32")
predictions
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['binary_accuracy'], label = 'train')
plt.show()
```

```python
plt.plot(history.history['loss'], label = 'train')
plt.show()
```
